VADER SENTIMENT SCORING

1.   We will be using NLTK's SentimentIntensityAnalyzer to get neg/neu/pos scores of the text.
2. using the "Bag of words" approach
     * stop words are removed
     * each word is scored and combomed to a total score.








```python
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer

from tqdm.notebook import tqdm

sia = SentimentIntensityAnalyzer()
```

```python
sia
```

```python
sia.polarity_scores('I am so happy')
```

```python
# run the polarity score on entire data set
sia = SentimentIntensityAnalyzer()
res = {}

for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Reviews']
    polarity_score1 = sia.polarity_scores(text)
    res[i] = polarity_score1
```

```python
res
```

```python
vaders = pd.DataFrame(res).T # (.T function flips the dataframe vertically)
```

```python
vaders
```

```python
vaders = vaders.merge(df, how='left', left_index=True, right_index=True)
```

```python
# now we have sentiment score and meta data
vaders.head()
```

```python
plot_r_c = sns.barplot(data=vaders, x='Ratings_out_of_5', y='compound')
plot_r_c.set_title('Compound Score by Users Ratings Review')
plt.show()
```

```python
# Create subplots for each sentiment score
fig, axes = plt.subplots(1, 3, figsize=(12, 3))

# Plot the barplots for each sentiment score
sns.barplot(data=vaders, x='Ratings_out_of_5', y='pos',  ax=axes[0])
sns.barplot(data=vaders, x='Ratings_out_of_5', y='neu',  ax=axes[1])
sns.barplot(data=vaders, x='Ratings_out_of_5', y='neg',  ax=axes[2])

# Set titles for each subplot
axes[0].set_title('Positive')
axes[1].set_title('Neutral')
axes[2].set_title('Negative')

plt.show()
```
