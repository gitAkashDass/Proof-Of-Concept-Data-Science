# This Is a Proof Of Concet Getting Data From Yfinance
* Sp500 Data was scraped and saves as a csv file for Further Analysis .
* BLACK ROCK INC Data was scraped and saves as a csv file for Further Analysis.

```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y='Close',use_index=True)
```

```python
sp500.to_csv('sp500_stock_data.csv')
```

```python
del sp500['Dividends']
del sp500['Stock Splits']
```

```python
Black_rock = yf.Ticker("BLK")
```
